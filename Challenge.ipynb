‚ùì Let's use [`seaborn.FacetGrid`](https://seaborn.pydata.org/generated/seaborn.FacetGrid.html)

1. Run the cell below. What do you observe?
2. Change `col` in the first line with another column (e.g. `"time"`). Run the cell again. What do you observe?

````python
import seaborn as sns
tips_df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
g = sns.FacetGrid(tips_df, col="day")
g.map(plt.hist, "total_bill")
````

The command above generates a facet histogram data visualization that plots the ```total_bill``` variable across different days of the week (Sunday, Saturday, Thursday, and Friday). This visualization allows us to analyse the density and frequency of the total bill amounts, providing insights into the profitability of the business on different days. Notably, there is a noticeable distinction between weekends and weekdays in terms of revenue generation. The visualization clearly shows that the business tends to generate more income during weekends compared to weekdays. This observation can be attributed to the higher number of customers visiting the establishment on weekends, likely due to it being a non-working day for the general public.

````python
g = sns.FacetGrid(tips_df, col="time")
g.map(plt.hist, "total_bill")
````

By analyzing this visualization, we can clearly distinguish the contrasting outcomes between ```Dinner``` and ```Lunch```. Notably, there is a substantial disparity in capital generation between the two, with dinner surpassing lunch. Several factors can be inferred as potential causes for this discrepancy. Firstly, the business might offer exclusive deals or promotions specifically tailored for dinner, thereby attracting more customers and generating higher revenue during that time. Alternatively, it is possible that the general public tends to have more leisure time later in the day, leading to increased patronage during dinner hours.

````python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

tips_df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
g = sns.FacetGrid(tips_df, col="day", row="smoker")
g.map(plt.hist, "total_bill")
````

8 cells were generated.

## Correlation

Let's start looking for correlation between columns in the dataset.


---
‚ùì What is your intuition about the relationship between the columns `tip` and `total_bill`?

The higher the bill the greater the tip.

---
‚ùì Let's look at the data to see if our intuition is correct. We will do a **scatterplot** with `x` being `total_bill` and `y` the tip.

````python
with sns.axes_style(style="whitegrid"):
    sns.relplot(x="total_bill", y="tip", data=tips_df)
````

---
‚ùì Another way of looking at this data is to use a [`seaborn.jointplot`](https://seaborn.pydata.org/generated/seaborn.jointplot.html).

````python
with sns.axes_style("white"):
    sns.jointplot(x="total_bill", y="tip", kind="hex", data=tips_df)
````

‚ùì A very useful tool to **identify** correlations is the [`seaborn.pairplot`](https://seaborn.pydata.org/generated/seaborn.pairplot.html):

````python
sns.pairplot(tips_df, height=2, hue="smoker")
````

## Regression

We are not doing Machine Learning yet but we can use [`seaborn.lmplot`](https://seaborn.pydata.org/generated/seaborn.lmplot.html) to graphically read a linear correlation between two columns:

````python
sns.lmplot(x="total_bill", y="tip", col="smoker", data=tips_df)
````
